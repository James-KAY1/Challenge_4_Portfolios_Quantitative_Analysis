##### Portfolio 1 & Portfolio 2- Step 1: Using the 60-day rolling window, the daily return data, and the S&P 500 returns, calculate the covariance. Review the last five rows of the covariance of the portfolio.

```python
# Calculate the covariance using a 60-day rolling window used Tiger Global Mngt LLC
# Review the last five rows of the covariance data
# YOUR CODE HERE
tgm_covariance_vs_sp = daily_return_whales_df["TIGER GLOBAL MANAGEMENT LLC"].rolling(window=60).cov(daily_returns_of_whales_nav_vs_sp["S&P 500"])
display(tgm_covariance_vs_sp.tail())
```

```python
# Calculate the covariance using a 60-day rolling window used Berkshire Hathaway Inc.
# Review the last five rows of the covariance data
# YOUR CODE HERE
bh_covariance_vs_sp = daily_return_whales_df["BERKSHIRE HATHAWAY INC"].rolling(window=60).cov(daily_returns_of_whales_nav_vs_sp["S&P 500"])
display(bh_covariance_vs_sp.tail())
```

##### Portfolio 1 & Portfolio 2 - Step 2: Calculate the beta of the portfolio. To do that, divide the covariance of the portfolio by the variance of the S&P 500.

```python
# Calculate the beta based on the 60-day rolling covariance compared to the market (S&P 500)
# Review the last five rows of the beta information
# YOUR CODE HERE
tgm_beta= tgm_covariance_vs_sp / market_rolling_60day_variance
bh_beta= bh_covariance_vs_sp /market_rolling_60day_variance

print ("Tiger Global Mngt LLC & Berkshire Hathaway 60 day rolling Beta respectively")
display(tgm_beta.tail())

display(bh_beta.tail())
```

##### Portfolio 1 & Porfolio 2 - Step 3: Use the Pandas `mean` function to calculate the average value of the 60-day rolling beta of the portfolio.

```python
# Calculate the average of the 60-day rolling beta
# YOUR CODE HERE
avg_tgm_beta = tgm_beta.mean()
avg_bh_beta = bh_beta.mean()
print(avg_tgm_beta)
print(avg_bh_beta)
```

##### Portfolio 1 & Portfolio 2- Step 4: Plot the 60-day rolling beta. Be sure to include the `title` parameter, and adjust the figure size if necessary.

```python
# Plot the rolling beta 
# Include a title parameter and adjust the figure size
# YOUR CODE HERE
ax = tgm_beta.plot(figsize= (10,7), title= " The Avg. 60 day rolling Beta for Tiger Global Mngt LLC & Berkshire Hathaway")
bh_beta.plot(ax=ax)
ax.legend(["Tiger Global Mgnt LLC", "Berkshire Hathaway"])
```
